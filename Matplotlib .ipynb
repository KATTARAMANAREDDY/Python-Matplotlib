Matplotlib Bar Plot

Matplotlib Scatter Plot

Matplotlib Pie Plot

Matplotlib Area Plot

# Matplotlib Line Plot

```python
# x-axis values

x = [2,8,6,9,3]

# y-axis values

y = [5,7,9,1,4]

# function to plot
plt.plot(x,y)

# function to show the plot
plt.show()
```

# Matplotlib Bar Plot

```python
# x-axis values

x = ['A','B','C','D','E']

# y-axis values

y = [5,10,15,20,25]

# function to plot
plt.bar(x,y)

# function to show the plot
plt.show()
```

# Matplotlib Histogram Plot

```python
# y-axis values

y = [2,8,6,9,3]

# function to plot
plt.hist(y)

# function to show the plot
plt.show()
```

# Matplotlib Scatter Plot

```python
# x-axis values

x = [2,4,6,8,10,8]

# y-axis values

y = [3,5,7,6,1,4]

# function to plot
plt.scatter(x,y)

# function to show the plot
plt.show()
```

# Matplotlib Pie Plot

```python
# Data

labels = ['M.P.C','Bi.P.C','C.E.C','M.E.C']

sizes = [37,39,40,33]

# function to plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Of Class 11 Students')

# function to show the plot
plt.show()
```

# Matplotlib Area Plot

```python
# Data

x = [2,4,6,8,10]

y1,y2 = [10,8,14,12,18,],[22,20,26,24,30]

# function to plot
plt.fill_between(x,y1,y2, color='brown', alpha=0.4, label='Area 1-2')
plt.plot(x,y1, label='Line 1', marker='o')
plt.plot(x,y2, label='Line 2', marker='o')

plt.title('Area chart Example')
plt.legend()

# function to show the plot
plt.show()
```
